## Compare step sizes: `9/π` vs `2π/9`

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
plot_unit_circle_with_steps(
    num_steps=36,
    step=DEFAULT_STEP_RADIANS,
    ax=axes[0],
    show_vortex_overlay=False,
    title="step = 9/π (dense / irregular)",
    label_first=8,
)
plot_unit_circle_with_steps(
    num_steps=9,
    step=2 * np.pi / 9,
    ax=axes[1],
    show_vortex_overlay=False,
    title="step = 2π/9 (equal 9-gon)",
    label_first=9,
)
plt.tight_layout()
plt.show()
```
